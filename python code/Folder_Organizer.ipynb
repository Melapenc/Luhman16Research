```python
# The file name with the specific word we are looking for from the long list of files.
desired_var = "bcd"

file_num = 1
n_index = file_num-1
# print(n_index)
limit_num = 8
```

```python
new_dir_arr=[]
for n_index in range(limit_num):
    num = [1,2,3,4,5,6,7]
    if n_index == 5:
        part_inp = str(6)
        part_num ="_part_"+part_inp+"-1"
    elif n_index == 6:
        part_inp = str(6)
        part_num ="_part_"+part_inp+"-2"
    elif n_index == 7:
        part_inp = str(7)
        part_num ="_part_"+part_inp
    else:
        part_inp = str(num[n_index])
        part_num ="_part_"+part_inp
    des_var_dir = "only_"+ desired_var + part_num
    created_dir = "/Users/melaniapena/Rsrch/Luhman_16_Research/bcd_files/" 
    save_dir = created_dir + des_var_dir
    new_dir_arr.append(save_dir) 
    
    print("Created directory:",save_dir)
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
        new_dir_arr.append(save_dir)
# new_dir_arr

# ----------------------------------------------------------------------------------------------------
#  This is a generic definition that automaticall creates folders with mor folders 
# within them and can put however many you want.
# def create_directories(Origin_dir, folder_name, number_of_folders):
#     new_dir_arr=[]
#     created_dir = Origin_dir
#     name_of_file_last = folder_name
#     val_range = number_of_folders
#     numb = np.linspace(1,val_range-1,val_range-1)
#     numb = np.append(numb,[val_range])
#     for n_index in range(val_range):
#         part_inp = str(int(numb[n_index-1]))
#         part_num =name_of_file_last+part_inp
#         save_dir = created_dir + part_num
#         new_dir_arr.append(save_dir) 
#         if not os.path.exists(save_dir):
#             os.makedirs(save_dir)
#             new_dir_arr.append(save_dir)
#     return new_dir_arr

# origin_dir = '/Users/melaniapena/Rsrch/Luhman_16_Research/test/'
# folder_name = 'testfolder_part_'

# test_1 = create_directories(origin_dir,folder_name,limit_num)
```

```python
## This for-loop only works when there's a list of some similar file names within 
## a folder already given to you and you desire to pick out those specific
## files and place them into a seperate new folder 
## (In our, case we are saving the new files into a new foler. Hence 'save_dir'.)

for n_index in range(limit_num):
    if os.path.exists(new_dir_arr[n_index]):
        for file_a in os.listdir(dir_BCD[n_index]):
##     joins the directory and file name again
            full_file_path = os.path.join(dir_BCD[n_index], file_a)
##     We are looking for file names that uses bcd fits.        
            str_file = file_a
##     creates the text file name as an array and is split depending on the given delimiter
##     For this file name.. The array is split into 8 indexes of an array.
##     Ex1: File name array: ['SPITZER', 'I1', '50037760', '0000', '0000', '2', 'bcd', 'fits']
            replacements = ('.', '_')
            for r in replacements:
                str_file = str_file.replace(r, ' ')
            str_file = str_file.split()
##     opens though the file name array and looks for the specific word you desired.
            for i in str_file:
##         Sets the desired word into a string. 
##         and checks if the file name has the word you wanted.
                str_i = str(i)
                if str_i == desired_var:
##            Joins the full directory together again
                    full_save_file_path = os.path.join(new_dir_arr[n_index], file_a)  
##            Checks if the file itself  already contains the same data. 
##            If it is, then there is no need to do anything else.
                    if os.path.exists(full_save_file_path) and os.path.getsize(full_save_file_path) >= 0:
                        print("This file already exists and contains the information needed.\nFile Name:",file_a,"\n\n")
##            Open the fits file from original directory and 
##            saves it into the new directoy we just 
##            created (which is full_save_file_path).
                    else:
                        hdulist = fits.open(full_file_path)
##            saves the data from the previous into our new file.
                        hdulist.writeto(full_save_file_path)
##            Then closes the fits file.
                        hdulist.close() 
                        print("New file created!\nFile Name:",file_a,"\n\n")
    else:
        print("This folder does not exists.\nYou must create the folder first")
    print("-------------------------------------------------------------\nProgram is done running.")
```
